```python
import pandas as pd
import glob
import os

# The path of the csv.
path = r'/Users/earltavera/Desktop/MSE803/Week 2/Week2A1/beijing+multi+site+air+quality+data/PRSA_Data_20130301-20170228'

# 1. Load the dataset (Merging all CSVs)
all_files = glob.glob(os.path.join(path, "*.csv"))
df_list = [pd.read_csv(f) for f in all_files]
df = pd.concat(df_list, ignore_index=True)

# 2. Display the first 5 rows
print("--- First 5 Rows ---")
print(df.head())

# 3. Identify column names and data types
print("\n--- Column Names and Data Types ---")
print(df.dtypes)

# 4. Count total rows and columns
rows, cols = df.shape
print(f"\n--- Dimensions ---")
print(f"Total Rows: {rows}")
print(f"Total Columns: {cols}")
```
